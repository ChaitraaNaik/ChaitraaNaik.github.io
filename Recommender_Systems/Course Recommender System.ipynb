```python
# code
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
import re
from spacy_langdetect import LanguageDetector
# import spacy
from textblob import TextBlob

import warnings
import neattext.functions as nfx
warnings.simplefilter(action='ignore', category=FutureWarning)
# df = pd.read_csv('usergoalsjoinlearn-edited2.csv',encoding='latin1')
# df = pd.read_csv('Gigturbodataset.csv',encoding='latin1')
# df = pd.read_csv('Gigturbodataset.csv',encoding='latin1')
df = pd.read_excel("coursera.xlsx",index_col=0)
# df containes lot of unnamed columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# df['id'] = [x for x in range(1, len(df.values)+1)]
df.dropna(inplace = True)
df['course_name'] = df['course_name'].apply(nfx.remove_stopwords)
df['course_name'] = df['course_name'].apply(nfx.remove_special_characters)
#df.index.name = 'id'
df
```

```python
#filtering out non english rows


from langdetect import detect

df_new = df[df.course_name.apply(detect).eq('en')]
df_new
```

```python
#Import TfIdfVectorizer from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

#Define a TF-IDF Vectorizer Object. Remove all english stop words such as 'the', 'a'
tfidf = TfidfVectorizer(stop_words='english')

#Replace NaN with an empty string
df_new['course_name'] = df_new['course_name'].fillna('')

def remove_pipeline(text):
    return re.sub("\\|", '', text)
# preg_match("/\|/", "This is not a pipe.

df_new['course_name']=df_new['course_name'].apply(remove_pipeline)
df_new['course_name']

#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(df_new['course_name'])

# #Output the shape of tfidf_matrix
tfidf_matrix.shape
```

```python
#Array mapping from feature integer indices to feature name.
tfidf.get_feature_names()[18:30]
```

```python
# Import linear_kernel
from sklearn.metrics.pairwise import cosine_similarity

# Compute the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
df_new.index = pd.RangeIndex(len(df_new.index))

df_new.reset_index(inplace=True)
df_new
```

```python
#Construct a reverse map of indices and course titles
indices = pd.Series(df_new.index, index=df_new['course_name']).drop_duplicates()
```

```python
# Function that takes in course title as input and outputs most similar courses
def get_recommendations(course_name, cosine_sim=cosine_sim):
    # Get the index of the course that matches the title
    idx = indices[course_name]

    # Get the pairwsie similarity scores of all courses with that course
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort the course based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    #print(sim_scores)
    # Get the scores of the 10 most similar courses
    sim_scores = sim_scores[1:15]

    # Get the course indices
    course_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar course
    return df_new['course_name'].iloc[course_indices]


# def recommend_course(title,num_of_rec=10):  
#     # Get the index of the course that matches the title
#     idx =  course_indices[title]

#     # Get the pairwsie similarity scores of all courses with that course
#     sim_scores = list(enumerate(cosine_sim_mat[idx]))

#     # Sort the course based on the similarity scores
#     sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

#     # Get the scores of the 10 most similar courses
#     sim_scores = sim_scores[1:10]

#     # Get the course indices
#     selected_course_indices = [i[0] for i in sim_scores]

#     # Return the top 10 most similar course
#     return df['course_name'].iloc[selected_course_indices]
df_new['course_name'].iloc[1474:]
```
